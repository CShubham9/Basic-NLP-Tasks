```python
import nltk
from nltk.corpus import brown
```

```python
#fiction category of brown corpus stored in b
b = brown.words(categories='fiction')
```

```python
b
```

```python
#collection of english stop words in s
from nltk.corpus import stopwords
s=set(stopwords.words('english')) 
```

```python
k=[]
for w in b:
    w=w.lower()
    w.append(k)
k
```

```python
#storing lower case equivalents of b 
k=[]
for w in b:
    w=w.lower()
    k.append(w)
```

```python
#removing stopwords from k
for w in k:
    if w in s:
        k.pop(k.index(w))
```

```python
k
```

```python
#removing punctuation marks from k
l=[',','.','``','--',"''",'?',';',':']
for w in k:
    if w in l:
        k.pop(k.index(w))
```
